### Emulating Computer Experiments through Hierarchical Dynamic Linear Models

In this notebook, I walk through simulated analysis of a hierarchical dynamic linear model (HDLM) and semingly unrelated time series equations (SUTSE) model. These can be used to link together multiple time series. In the GaSP framework, this linking is achieved through the GASP covariance function applied to the space of simulation inputs. For example, the Liu-West dynamic emulation framework seeks to build a fast surrogate model of an expensive computer simulator that outputs long and possibly nonstationary time series. This is achieved through the use of a 1-level hierarchical DLM. Let the state-space model be formulated as follows: for $m$ simulator runs, let $\mathbf y_t\in\mathbb R^{m}$ be the collection of computer outputs so that

$$
\begin{aligned}
\mathbf y_t = F_t\boldsymbol\theta_{t} + v_t,\ v_t\sim\mathcal N(0, \sigma_t\Sigma)\\
\boldsymbol\theta_{t} =\boldsymbol\theta_{t-1} + w_t,\ v_t\sim\mathcal N(0, W_t).
\end{aligned}
$$

In this formulation, the measurement transition matrix $F_t$ is allowed to be a $m\times p$ dimensional autoregressive matrix, were the columns correspond to a specific lag. For a concrete example, let $F_t$ be an AR(1) matrix with lags equal to $\mathbf y_{t-1}$. Then $\boldsymbol \theta_t\in\mathbb R$, and the "borrowing" is achieved by assuming that each series is drived by $F_t\boldsymbol\theta_t$ and all variances between are "absorbed" by the GaSP covariance. This covariance, $\Sigma$, induces this relevant correlation among the series based on computer input values through the Gaussian process kernel function. Finally, the relevant temporal correlations are captured through the process innovation governed through the covariance matrix $W_t$. Emulation and calibration of the above DLM formulation is then relatively straightforward, as all distributions are Gaussian and relevant quantites of interest are formed through the necessary posterior predictive distributions.

The extension to spatio-temporal data can take many forms, but a conceptually immediate formulation is through the work of (Landim and Gammerman, 2000), where a matrix-variate hierarchical DLM is formulated. To motivate this approach, imagine a square $n^2$ spatial grid of observations.

The matrix-variate formulation can be advantageous because the rows might contain a computer simulation replicate at a specific time point, while the columns contain a vectorized spatial grid. In this case, the hierarchical DLM takes the form

$$
\begin{aligned}
\mathbf Y_t = \mathbf F_t \boldsymbol\Theta_t + V_t, \ V_t\sim\text{MN}(\boldsymbol 0, \overbrace{\Sigma}^{GaSP}, \overbrace{\Omega_1\otimes\Omega_2}^{Spatial\ Structure})\\
\boldsymbol\Theta_t = \boldsymbol\Theta_{t-1} + W_t, \ V_t\sim\text{MN}(\boldsymbol 0, \Lambda_1, \Lambda_2).
\end{aligned}
$$

Now the $\mathbf F_t$ is the measurement transition matrix is taken to represent the evolution of the matrix-variate process

### Experiments with Hierarchical DLMs in Stan
The following are some notes and experiments with higher-order hierarchical DLMs implemented in Stan. For now, they are univariate, but it's through the development of these that I will work towards the matrix-normal implementation in Stan. There is no GaSP component to these models -- yet.

$$
\begin{aligned}
Y_t &= F_{t}\theta_t + v_t, \ v_t\sim\mathcal N_m(0,V_t)\\
\theta_t &= H_{t}\lambda_t + u_t, \ u_t\sim\mathcal N_p(0,U_t)\\
\lambda_t &= G_t\lambda_{t-1} + w_t, \ w_t\sim\mathcal N_k(0,W_t)
\end{aligned}
$$

One application of this hierarchical structure is to connect multiple time series. For instance, consider a collection of $m$ time series recored at time $t$ so that $Y_t = (Y_{1,t},\ldots,Y_{m,t})^\top$. Let each individual series evolve according to a univariate DLM so that

$$
Y_{i,t} = F_{i,t}\theta_{i,t} + \varepsilon_t, \ \varepsilon_t\sim\mathcal N_m(0,\sigma_{i,t}^2).
$$

In this case, there could be as many as $m\times p \times t$ number of process parameters to estimate which can be enormous. Instead, a HDLM can be employed to link the process parameters together by assuming a commont latent structure such that

$$
\begin{aligned}
\theta_{i,t} &= H_t\lambda_t + \tau_{i,t}, \ \tau_{i,t}\sim\mathcal N_p(0,\Sigma_t)\\
\lambda_t &= G_t\lambda_{t-1} + w_t, \ w_t\sim\mathcal N_k(0,W_t).
\end{aligned}
$$

In this way, the number of process paramters has been reduced from $m\times p \times t$ to $p \times t$. To write this in full HDLM form, the neccesary components of (?) are stacked into block vectors and matricies so that

$$
\begin{aligned}
\begin{pmatrix}
Y_{1,t}\\
\vdots\\
Y_{m,t}
\end{pmatrix}&=
\begin{pmatrix}
F_{1,t}& &\\
& \ddots &\\
& & F_{1,t}
\end{pmatrix}
\begin{pmatrix}
\theta_{1,t}\\
\vdots\\
\theta_{m,t}
\end{pmatrix} + v_t, \ v_t\sim\mathcal N_m(\boldsymbol 0,\text{diag}(\sigma_1^2,\ldots,\sigma_m^2))\\
\begin{pmatrix}
\theta_{1,t}\\
\vdots\\
\theta_{m,t}
\end{pmatrix}&=\begin{pmatrix}
I_p\\
\vdots\\
I_p
\end{pmatrix}\lambda_t + u_t, \ u_t\sim\mathcal N_{mp}(\boldsymbol 0,\text{diag}(\Sigma_t,\ldots,\Sigma_t))\\\\
\lambda_t &= G_t\lambda_{t-1} + w_t, \ w_t\sim\mathcal N_k(\boldsymbol 0, W_t)
\end{aligned}
$$

By colapsing the above hierarchy, the standard state-space formulation can be recovered:

$$
\begin{aligned}
Y_t &= F_tH_t\lambda_t + v^*_t, \ v^*_t\sim\mathcal N_m(\boldsymbol 0, F_t\text{diag}(\Sigma_t,\ldots,\Sigma_t))F_t^\top+\text{diag}(\sigma_1^2,\ldots,\sigma_m^2))\\
\lambda_t &= G\lambda_{t-1}+w_t, \ w_t\sim\mathcal N_p(\boldsymbol 0, W_t).
\end{aligned}
$$

```python
# Import data block
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import scipy
from cmdstanpy import cmdstan_path, CmdStanModel
import arviz as az
```

```python
#####
# Hierarchical DLM
#####

m, p = 3, 3
T = 1000
y = np.zeros((T,m))
F = np.ones((1, p))
Ft = np.zeros((m, m*p , T))
G = np.eye(p)

W = .1*np.eye(p) # datasets.make_spd_matrix(n_dim=p)
Sigma = .1*np.eye(p)
v = .2*np.eye(m)

# generate big state transition matirx for t
y[1,:] = np.random.multivariate_normal(np.zeros(p), W, 1)
for t in range(1,T):
    # uncomment for autoregressive model
    Ft[:, :, t] = np.kron(np.eye(m), y[t-1,:])
    # uncomment for fixed state transition matrix
    #Ft[:, :, t] = np.kron(np.eye(m), np.ones((1,p)))

    y[t,:] = np.random.multivariate_normal(np.zeros(p), W, 1)

H = np.concatenate([np.eye(p) for x in range(0,m)]) 

# generate fake process variables
lambda_ = np.zeros((T,p))
lambda_[1,:] = np.random.multivariate_normal(np.zeros(p), W, 1)
phi = 1*np.ones(T)
for t in range(1,T):
    lambda_[t,:] = phi[t] * lambda_[t-1] + np.random.multivariate_normal(np.zeros(p), W, 1)
    # uncomment for autoregressive mmodel
    #phi[t] = scipy.random.normal(phi[t-1],.5) 
    
# generate fake observation process
Sigma_diag = np.zeros((p*m, p*m, T))
Sig = 1*np.eye(p)
for t in range(0,T):
    Sigma_diag[:,:,t] = np.kron(np.eye(p, dtype=int), Sig)
    y[t,:]=np.transpose(Ft[:,:,t] @ H @ lambda_[t,:]) \
    + np.random.multivariate_normal(np.zeros(m), \
                                    Ft[:,:,t] @ Sigma_diag[:,:,t] @ np.transpose(Ft[:,:,t]) + v, 1)
plt.plot(y)
```

An alternative to the hierarchical DLM from the above section is the *Seemingly Unrelated Time Series Equations* (SURTE) model. This might be approprioate when there's too much "borrowing" of information across series in the hierarchical model. Consider again $m$ univeriate time series. In the case of spatio-temporal data, these may be obtained through the use of the Kronecker product of the relevant covariance structures

```python
W_mu = [[1, .5],[.5,1]]
W_beta = [[1, .6],[.6,1]]
# W_mu = datasets.make_spd_matrix(n_dim=2)
# W_beta = datasets.make_spd_matrix(n_dim=2)
W = 1e-5*scipy.linalg.block_diag(W_mu,W_beta)
```

```python
####
# Seemingly Unrelated Time Series Equations (SUTES)
T = 250
G = [[1,0,1,0],
     [0,1,0,1],
     [0,0,1,0],
     [0,0,0,1]]
F = [[1,0,0,0], 
    [0,1,0,0]]
V = .1 * np.eye(2)
theta = np.zeros((T, 4))
y = np.zeros((T,2))
theta[1,:] = np.random.multivariate_normal(np.zeros(4), W)
for t in range(1,T):
    theta[t, :] =  np.random.multivariate_normal(G @ theta[t-1,:], W)
for t in range(0,T):
    y[t, :] = F @ theta[t, :] + np.random.multivariate_normal(np.zeros(2), V)
    
plt.plot(y)
```

```python
df = {'y': y, 'N': y.shape[0], 'F': F, 'G': G}
model = CmdStanModel(stan_file="sutse.stan")
```

```python
samples = model.sample(data=df,max_treedepth = 10,adapt_delta = .8,
                             chains = 2, 
                             iter_warmup = 150,
                             iter_sampling = 150,show_progress=True,
                             refresh = 50, parallel_chains = 4)
```

```python
azsamples = az.from_cmdstanpy(samples)
s = samples.stan_variable("ypred")
plt.plot(s.mean(axis=0))
plt()
```
